```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Datasets
features = pd.read_csv("Features data set.csv")
sales = pd.read_csv("sales data-set.csv")
stores = pd.read_csv("stores data-set.csv")
```

```python
# Merge Datasets
merged_data = pd.merge(sales, features, on=['Store', 'Date'], how='left')
merged_data = pd.merge(merged_data, stores, on='Store', how='left')

# Convert Date column to datetime format
merged_data['Date'] = pd.to_datetime(merged_data['Date'])
```

```python
# Check for missing value
print("Missing values in Merged dataset:")
print(merged_data.isnull().sum())
```

```python
# Calculate KPIs
total_sales = merged_data['Weekly_Sales'].sum()
average_sales = merged_data['Weekly_Sales'].mean()
average_temperature = merged_data['Temperature'].mean()
average_cpi = merged_data['CPI'].mean()
average_unemployment = merged_data['Unemployment'].mean()
```

```python
# Display KPIs
print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)
print("Average Temperature:", average_temperature)
print("Average CPI:", average_cpi)
print("Average Unemployment Rate:", average_unemployment)
```

```python
# Visualizations
plt.figure(figsize=(15, 8))

# Line plot for Weekly Sales over time
plt.subplot(2, 2, 1)
sns.lineplot(x='Date', y='Weekly_Sales', data=merged_data, ci=None)
plt.title('Weekly Sales Over Time')

# Boxplot for Weekly Sales by Store Type
plt.subplot(2, 2, 2)
sns.boxplot(x='Type', y='Weekly_Sales', data=merged_data)
plt.title('Weekly Sales by Store Type')

# Scatter plot for Temperature vs. Weekly Sales
plt.subplot(2, 2, 3)
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=merged_data)
plt.title('Temperature vs. Weekly Sales')

# Histogram for Unemployment Rate
plt.subplot(2, 2, 4)
sns.histplot(merged_data['Unemployment'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate')

plt.tight_layout()
plt.show()
```

```python
# Merge Datasets
merged_data = pd.merge(sales, features, on=['Store', 'Date'], how='left')
merged_data = pd.merge(merged_data, stores, on='Store', how='left')

# Convert Date column to datetime format
merged_data['Date'] = pd.to_datetime(merged_data['Date'])
```

```python
# Check for missing values
print("Missing values in Merged dataset:")
print(merged_data.isnull().sum())

# Calculate KPIs
total_sales = merged_data['Weekly_Sales'].sum()
average_sales = merged_data['Weekly_Sales'].mean()
average_temperature = merged_data['Temperature'].mean()
average_cpi = merged_data['CPI'].mean()
average_unemployment = merged_data['Unemployment'].mean()
total_stores = merged_data['Store'].nunique()
average_store_size = stores.groupby('Type')['Size'].mean()
```

```python
# Display KPIs
print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)
print("Average Temperature:", average_temperature)
print("Average CPI:", average_cpi)
print("Average Unemployment Rate:", average_unemployment)
print("Total Number of Stores:", total_stores)
print("Average Store Size by Type:\n", average_store_size)
```

```python
# Visualizations
plt.figure(figsize=(18, 12))

# Line plot for Weekly Sales over time
plt.subplot(3, 2, 1)
sns.lineplot(x='Date', y='Weekly_Sales', data=merged_data, ci=None)
plt.title('Weekly Sales Over Time')

# Boxplot for Weekly Sales by Store Type
plt.subplot(3, 2, 2)
sns.boxplot(x='Type', y='Weekly_Sales', data=merged_data)
plt.title('Weekly Sales by Store Type')

# Scatter plot for Temperature vs. Weekly Sales
plt.subplot(3, 2, 3)
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=merged_data)
plt.title('Temperature vs. Weekly Sales')

# Histogram for Unemployment Rate
plt.subplot(3, 2, 4)
sns.histplot(merged_data['Unemployment'], bins=20, kde=True)
plt.title('Distribution of Unemployment Rate')

# Bar plot for Average Store Size by Type
plt.subplot(3, 2, 5)
sns.barplot(x=average_store_size.index, y=average_store_size.values)
plt.title('Average Store Size by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Average Store Size')

# Line plot for CPI over time
plt.subplot(3, 2, 6)
sns.lineplot(x='Date', y='CPI', data=merged_data, ci=None)
plt.title('CPI Over Time')

plt.tight_layout()
plt.show()
```

```python
# Create a pair plot with hue based on holiday
sns.pairplot(merged_data, hue='IsHoliday_x', diag_kind='kde', height=3)
plt.suptitle('Pairplot for Numerical Variables with Holiday Highlight', y=1.02)
plt.show()
```

```python
# Heatmap for correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = merged_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Violin plot for Weekly Sales distribution by Store Type
plt.figure(figsize=(12, 6))
sns.violinplot(x='Type', y='Weekly_Sales', data=merged_data, inner='quartile')
plt.title('Weekly Sales Distribution by Store Type')
plt.xlabel('Store Type')
plt.ylabel('Weekly Sales')
plt.show()
```
